```python
import os
import matplotlib.pyplot as plt
import SimpleITK as sitk
import random
```

```python
root_dir = "./Task09_Spleen/Task09_Spleen/"
```

```python
# Number of Training and Testing cases 
train_dir = os.path.join(root_dir, "imagesTr")
test_dir = os.path.join(root_dir, "imagesTs")
num_train_cases = len(os.listdir(train_dir))
num_test_cases = len(os.listdir(test_dir))
print(f"Number of Training Cases: {num_train_cases}")
print(f"Number of Testing Cases: {num_test_cases}")
```

```python
# Check Characterstics of Training Cases 
train_cases = os.listdir(train_dir)
train_cases = list(filter(lambda x: not x.startswith("._"), train_cases))
print(f"First 5 Training Cases: {train_cases[:5]}")

# Load 5 random samples and check their dimensions and spacing
random_samples = random.sample(train_cases, 5)
for case in random_samples:
    img = sitk.ReadImage(os.path.join(train_dir, case))
    print(f"Case: {case}")
    print(f"Dimensions: {img.GetDimension()}")
    print(f"Size: {img.GetSize()}")
    print(f"Spacing: {img.GetSpacing()}")
    print("---")
```
