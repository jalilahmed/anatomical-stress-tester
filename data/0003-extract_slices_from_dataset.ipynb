```python
import os
import SimpleITK as sitk
import numpy as np
```

```python
root_dir = "./Task09_Spleen/Task09_Spleen/"
```

```python
# Number of Training and Testing cases 
train_dir = os.path.join(root_dir, "imagesTr")
test_dir = os.path.join(root_dir, "imagesTs")
num_train_cases = len(os.listdir(train_dir))
num_test_cases = len(os.listdir(test_dir))
print(f"Number of Training Cases: {num_train_cases}")
print(f"Number of Testing Cases: {num_test_cases}")
```

```python
# Training cases 
for train_case in os.listdir(train_dir):
    train_case_path = os.path.join(train_dir, train_case)
    label_case_path = os.path.join(root_dir, "labelsTr", train_case.replace("image", "label"))
    
    # Read the image and label
    image = sitk.ReadImage(train_case_path)
    label = sitk.ReadImage(label_case_path)
    
    # Convert to numpy arrays
    image_array = sitk.GetArrayFromImage(image)
    label_array = sitk.GetArrayFromImage(label)

    # Asses images and labels array have same shape 
    assert image_array.shape == label_array.shape, f"Image and label shapes do not match for {train_case}"
    
    # Iterate through slices and save those with labels
    for slice_index in range(image_array.shape[0]):
        if label_array[slice_index].sum() > 50:  # Check if there are any labels in the slice and greater than threshold
            slice_image = image_array[slice_index]
            slice_label = label_array[slice_index]
            
            # Save the slice and label as needed (e.g., using numpy or SimpleITK)
            np.save(f"../data/Task09_Spleen/slices/imagesTr/{train_case}_{slice_index}.npy", slice_image)
            np.save(f"../data/Task09_Spleen/slices/labelsTr/{train_case}_{slice_index}.npy", slice_label)
```

```python
# Testing cases 
for test_case in os.listdir(test_dir):
    test_case_path = os.path.join(test_dir, test_case)
    
    # Read the image and label
    image = sitk.ReadImage(train_case_path)

    # Convert to numpy arrays
    image_array = sitk.GetArrayFromImage(image)

    # Iterate through slices and save those with labels
    for slice_index in range(image_array.shape[0]):
        if label_array[slice_index].sum() > 50:  # Check if there are any labels in the slice and greater than threshold
            slice_image = image_array[slice_index]
            slice_label = label_array[slice_index]
            
            # Case basename without extension
            case_name = test_case.split('.')[0]
            
            # Save the slice and label as needed (e.g., using numpy or SimpleITK)
            np.save(f"../data/Task09_Spleen/slices/imagesTs/{case_name}_{slice_index}.npy", slice_image)
```
